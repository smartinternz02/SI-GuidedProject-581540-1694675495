```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the car crashes dataset (you can replace this with your dataset)
car_crashes = sns.load_dataset("car_crashes")
```

```python
# Plot 1: Pairplot
sns.pairplot(car_crashes)
plt.title("Pairplot of Car Crashes Data")
plt.show()

# Inference 1:
# The pairplot shows scatter plots of all numerical variables against each other.
# You can observe relationships between variables, such as speed vs. alcohol,
# and the distribution of each variable. For example, it appears that there
# may be a positive correlation between alcohol and total crashes.
```

```python
# Plot 2: Heatmap of Correlation
corr_matrix = car_crashes.corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
# Inference 2:
# The correlation heatmap visualizes the pairwise correlations between numerical
# variables. It helps identify strong positive or negative correlations.
# In this dataset, you can see that "total" and "alcohol" have a relatively strong
# positive correlation, which suggests that higher alcohol consumption might
# lead to more crashes.
```

```python
# Plot 3: Boxplot of Insurance Premiums
sns.boxplot(x="total", y="ins_premium", data=car_crashes)
plt.title("Boxplot of Total Crashes vs. Insurance Premiums")
plt.show()

# Inference 3:
# The boxplot shows the distribution of insurance premiums for different levels of
# total crashes. It helps identify potential outliers and the central tendency of
# insurance premiums for each group. From the plot, it seems that as the number
# of total crashes increases, the median insurance premium also tends to increase.
```

```python
# Plot 4: Distribution of Alcohol Consumption
sns.histplot(car_crashes["alcohol"], kde=True)
plt.title("Distribution of Alcohol Consumption")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Frequency")
plt.show()

# Inference 4:
# This histogram shows the distribution of alcohol consumption levels in the dataset.
# It helps you understand the central tendency and the spread of alcohol consumption
# among the recorded car crashes. You can see whether it follows a normal distribution
# or if there are any outliers.
```
